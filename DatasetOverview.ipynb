# Cat Dataset from Kaggle

```python
import opendatasets as od
dataset_url = 'https://www.kaggle.com/datasets/borhanitrash/cat-dataset/data'
od.download(dataset_url, data_dir='/content/datasets')
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Ścieżka do folderu z danymi
dataset_path = '/content/datasets/cat-dataset/cats/Data/'

# Znalezienie wszystkich ścieżek do obrazków
image_paths = []
for root, _, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(root, file))

# Wylosowanie 50 obrazków (lub mniej jeśli jest ich mniej)
num_images = min(60, len(image_paths))
random_images = random.sample(image_paths, num_images)

# Wyświetlenie w siatce 5x10
rows, cols = 6, 10
plt.figure(figsize=(18, 12))  # Dopasuj rozmiar figury
plt.title("\n Cat Dataset", fontsize = 24, fontweight = 'bold')
plt.axis('off')

for i, img_path in enumerate(random_images):
    img = Image.open(img_path)
    plt.subplot(rows, cols, i + 1)
    plt.imshow(img)
    plt.axis('off')
    # Można dodać nazwę pliku jako tytuł (opcjonalnie)
    # plt.title(os.path.basename(img_path), fontsize=8)


plt.tight_layout()
plt.show()
```
